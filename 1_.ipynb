```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemple de donnes

# Liste des motions uniques
emotions = df['Emotion'].unique()

# Calcul du pourcentage de chaque motion
emotion_percentages = df['Emotion'].value_counts(normalize=True) * 100

# Cration de la figure et des sous-graphiques
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Pie chart - Pourcentage de chaque motion
ax1.pie(emotion_percentages, labels=emotions, autopct='%1.1f%%')
ax1.set_title('Pourcentage de chaque motion')

# Dot plot - Pourcentage de chaque motion
ax2.plot(emotions, emotion_percentages, 'bo', markersize=10)
ax2.set_title('Pourcentage de chaque motion')
ax2.set_xlabel('motion')
ax2.set_ylabel('Pourcentage')

# Ajuster les espacements entre les sous-graphiques
plt.tight_layout()

# Afficher le graphique
plt.show()
```
