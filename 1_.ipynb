```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.pie(  ### PIE CHART
    x=df.Emotion.value_counts(normalize=True) * 100, 
    labels=df.Emotion.unique(), 
    autopct='%1.1f%%',
    startangle=-45,
    colors=['steelblue', 'tomato', 'hotpink', 'silver', 'yellowgreen', 'orange']
)
ax1.set_title('Représentation de chaque émotion dans le corpus')

ax2.plot( ### DOT PLOT
    df.Emotion.unique(), df.Emotion.value_counts(), 'o', color='black')

bars = ax2.bar(
    df['Emotion'].unique(), df['Emotion'].value_counts(), width=0.04, color=['steelblue', 'tomato', 'hotpink', 'silver', 'yellowgreen', 'orange'])
ax2.set_title('Taux pour chaque émotion')
ax2.set(xlabel='Émotion', ylabel='Nombre de lignes')

# Ajouter les valeurs annotées au-dessus de chaque barre
ax2.bar_label(bars, labels=df['Emotion'].value_counts().values, label_type='edge', padding=4)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (20,7))
#sns.set_theme(style="darkgrid")

ax = sns.countplot(x="Emotion", data=df, palette="Set3", dodge=False)
```
